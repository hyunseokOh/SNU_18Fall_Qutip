```python
from qutip import *
from qutip.ui.progressbar import TextProgressBar as ProgressBar
import numpy as np
import matplotlib.pyplot as plt

options=Options(nsteps=20000)

#  Find optimal frequency of microwave
def freqCalc(rho0,frelist,tlist,target):
    wmw1=0
    wmw2=0
    wmw3=0

    rmat=np.zeros((len(frelist),len(tlist)))
    pbar=ProgressBar(len(frelist))
    c_ops=[]

    if g1>0.0:
        c_ops.append(np.sqrt(g1)*sigmam())

    if g2>0.0:
        c_ops.append(np.sqrt(g2)*sigmaz())

    #  H0=(w01/2.0)*tensor(sigmaz(),qeye(2))+(w02/2.0)*tensor(qeye(2),sigmaz())+(J12/2.0)*tensor(sigmaz(),sigmaz())
    #  H1=Bac*tensor(sigmax(),qeye(2))
    #  H=[H0,[H1,'np.cos(wmw*t)']]
    
    sx=sigmax()
    sz=sigmaz()
    idm=qeye(2)
    
    for i in range(len(frelist)):
        pbar.update(i)
        
        if target==1:
            e_ops=[tensor(sz,idm,idm)]
            wmw1=frelist[i]
        
        elif target==2:
            e_ops=[tensor(idm,sz,idm)]
            wmw2=frelist[i]
            
        else:
            e_ops=[tensor(idm,idm,sz)]
            wmw3=frelist[i]

        H0=(w01-wmw1)/2*tensor(sz,idm,idm)+(w02-wmw2)/2*tensor(idm,sz,idm)+(w03-wmw3)/2*tensor(idm,idm,sz)+J12*tensor(sz,sz,idm)+J23*tensor(idm,sz,sz)
        Hint=(Bac1/4)*tensor(sx,idm,idm)+(Bac2/4)*tensor(idm,sx,idm)+(Bac3/4)*tensor(idm,idm,sx)
        H=H0+Hint
        
        output=mesolve(H,rho0,tlist,c_ops,e_ops,options=options)
        rmat[i]=output.expect[0]
    return rmat

def freqFind12(frelist,tlist):
    rho1=ket2dm(tensor(basis(2,0),basis(2,0),basis(2,0)))
    rmat1=freqCalc(rho1,frelist,tlist,1)

    rho2=ket2dm(tensor(basis(2,0),basis(2,1),basis(2,0)))
    rmat2=freqCalc(rho2,frelist,tlist,1)

    opt=np.where(rmat2==np.min(rmat2))
    f_opt=frelist[opt[0][0]]
    t_opt=tlist[opt[1][0]]
    
    fig,(ax1,ax2)=plt.subplots(1,2)
    t_mat,fre_mat=np.meshgrid(tlist,frelist/2/np.pi)
    c1=ax1.pcolor(t_mat,fre_mat,rmat1)
    c2=ax2.pcolor(t_mat,fre_mat,rmat2)
    fig.colorbar(c1,ax=ax1)
    fig.colorbar(c2,ax=ax2)
    plt.show()
    
    frelist=np.linspace(f_opt-(ff-fi)/25,f_opt+(ff-fi)/25,101)
    tlist=np.linspace(t_opt-(tf-ti)/25,t_opt+(tf-ti)/25,101)
    tlist=np.insert(tlist,0,[0])

    rmat2=freqCalc(rho2,frelist,tlist,1)
    opt=np.where(rmat2==np.min(rmat2))
    f_opt=frelist[opt[0][0]]
    t_opt=tlist[opt[1][0]]
    print("Optimal frequency : %f, Optimal time : %f" % (f_opt/2/np.pi,t_opt))
    print()
    
  
    
    return (f_opt,t_opt)


def freqFind23(frelist,tlist):
    rho1=ket2dm(tensor(basis(2,0),basis(2,0),basis(2,0)))
    rmat1=freqCalc(rho1,frelist,tlist,3)

    rho2=ket2dm(tensor(basis(2,0),basis(2,1),basis(2,0)))
    rmat2=freqCalc(rho2,frelist,tlist,3)

    opt=np.where(rmat2==np.min(rmat2))
    f_opt=frelist[opt[0][0]]
    t_opt=tlist[opt[1][0]]
    
    fig,(ax1,ax2)=plt.subplots(1,2)
    t_mat,fre_mat=np.meshgrid(tlist,frelist/2/np.pi)
    c1=ax1.pcolor(t_mat,fre_mat,rmat1)
    c2=ax2.pcolor(t_mat,fre_mat,rmat2)
    fig.colorbar(c1,ax=ax1)
    fig.colorbar(c2,ax=ax2)
    plt.show()

    frelist=np.linspace(f_opt-(ff-fi)/25,f_opt+(ff-fi)/25,101)
    tlist=np.linspace(t_opt-(tf-ti)/25,t_opt+(tf-ti)/25,101)
    tlist=np.insert(tlist,0,[0])

    rmat2=freqCalc(rho2,frelist,tlist,3)
    opt=np.where(rmat2==np.min(rmat2))
    f_opt=frelist[opt[0][0]]
    t_opt=tlist[opt[1][0]]
    print("Optimal frequency : %f, Optimal time : %f" % (f_opt/2/np.pi,t_opt))
    print()

    return (f_opt,t_opt)

def freqFind2(frelist,tlist):
    rho1=ket2dm(tensor(basis(2,0),basis(2,0),basis(2,0)))
    rmat1=freqCalc(rho1,frelist,tlist,2)

    opt=np.where(np.abs(rmat1)==np.min(np.abs(rmat1)))
    f_opt=frelist[opt[0][0]]
    t_opt=tlist[opt[1][0]]
    
    
    fig,(ax1)=plt.subplots()
    t_mat,fre_mat=np.meshgrid(tlist,frelist/2/np.pi)
    c1=ax1.pcolor(t_mat,fre_mat,rmat1)
  
    fig.colorbar(c1,ax=ax1)

    frelist=np.linspace(f_opt-(ff-fi)/25,f_opt+(ff-fi)/25,101)
    tlist=np.linspace(t_opt-(tf-ti)/25,t_opt+(tf-ti)/25,101)
    tlist=np.insert(tlist,0,[0])

    rmat1=freqCalc(rho1,frelist,tlist,2)
    opt=np.where(np.abs(rmat1)==np.min(np.abs(rmat1)))
    f_opt=frelist[opt[0][0]]
    t_opt=tlist[opt[1][0]]


    plt.show()
    print(np.min(np.abs(rmat1)))
    print("Optimal frequency : %f, Optimal time : %f" % (f_opt/2/np.pi,t_opt))
    print()
    return (f_opt,t_opt)

def Hadamard2_impl(rho0,f_opt,t_opt):
    wmw1=0
    wmw2=f_opt
    wmw3=0
    sx=sigmax()
    sz=sigmaz()
    idm=qeye(2)
    H0=(w01-wmw1)/2*tensor(sz,idm,idm)+(w02-wmw2)/2*tensor(idm,sz,idm)+(w03-wmw3)/2*tensor(idm,idm,sz)+J12*tensor(sz,sz,idm)+J23*tensor(idm,sz,sz)
    Hint=(Bac1/4)*tensor(sx,idm,idm)+(Bac2/4)*tensor(idm,sx,idm)+(Bac3/4)*tensor(idm,idm,sx)
    H=H0+Hint

#     tlist_rot=np.linspace(0,t_opt,100)
    tlist_rot=[0,t_opt]
    output=mesolve(H,rho0,tlist_rot)

#     output=mesolve(H,rho0,tlist_rot,[],[tensor(idm,sx,idm),tensor(idm,sigmay(),idm),tensor(idm,sz,idm)],options=options)
    
    # Rotation without microwave, remove phase shift
    rho1=output.states[-1]
    
    Hrot=w01/2*tensor(sz,idm,idm)+w02/2*tensor(idm,sz,idm)+w03/2*tensor(idm,idm,sz)+J12*tensor(sz,sz,idm)+J23*tensor(idm,sz,sz)
    

    tlist_rot=np.linspace(0,0.5*np.pi/J23,2)
    


    output=mesolve(Hrot,rho1,tlist_rot)
#     return output.expect[0],output.expect[1],output.expect[2]
    return output.states
    
def CNOT_impl(rho0,f_opt,t_opt,target):
    wmw1=0
    wmw2=0
    wmw3=0

    tl=[0,t_opt]
    c_ops=[]

    if g1>0.0:
        c_ops.append(np.sqrt(g1)*sigmam())

    if g2>0.0:
        c_ops.append(np.sqrt(g2)*sigmaz())
    
    if target==1:
        wmw1=f_opt
    else:
        wmw3=f_opt
    
    sz=sigmaz()
    sx=sigmax()
    idm=qeye(2)
    
    H0=(w01-wmw1)/2*tensor(sz,idm,idm)+(w02-wmw2)/2*tensor(idm,sz,idm)+(w03-wmw3)/2*tensor(idm,idm,sz)+J12*tensor(sz,sz,idm)+J23*tensor(idm,sz,sz)
    Hint=(Bac1/4)*tensor(sx,idm,idm)+(Bac2/4)*tensor(idm,sx,idm)+(Bac3/4)*tensor(idm,idm,sx)
    H=H0+Hint
    
    output=mesolve(H,rho0,tl,c_ops,[],options=options)
    
    # Rotation without microwave, remove phase shift
    rho1=output.states[-1]
    Hrot=w01/2*tensor(sz,idm,idm)+w02/2*tensor(idm,sz,idm)+w03/2*tensor(idm,idm,sz)+J12*tensor(sz,sz,idm)+J23*tensor(idm,sz,sz)
    
    if target==1:
        tlist_rot=np.linspace(0,1.5*np.pi/J12,2)
    
    elif target==3:
        tlist_rot=np.linspace(0,1.5*np.pi/J23,2)

    output=mesolve(Hrot,rho1,tlist_rot)
    
    return output.states

```

```python

#  Parameter settings
w01=11*2*np.pi
w02=12*2*np.pi
w03=10*2*np.pi
Bac1=0.2*2*np.pi
Bac2=0.2*2*np.pi
Bac3=0.2*2*np.pi
J12=1*2*np.pi
J23=2*2*np.pi
g1=0.0
g2=0.0

fi=0
ff=20
ti=0
tf=10

frelist=np.linspace(fi,ff,101)*2*np.pi
tlist=np.linspace(ti,tf,101)
```

```python
Bac2=Bac3=0
f1,t1=freqFind12(frelist,tlist)
Bac3=0.2*2*np.pi
Bac1=0
f3,t3=freqFind23(frelist,tlist)
Bac3=0
Bac2=0.2*2*np.pi
f2,t2=freqFind2(frelist,tlist)
```

```python
psi1=basis(2,0)
psi2=basis(2,0)
psi3=basis(2,0)
rho0=ket2dm(tensor(psi1,psi2,psi3))

Bac2=0.2*2*np.pi
Bac1=0
Bac3=0
rho1=Hadamard2_impl(rho0,f2,t2)[-1]

# x,y,z=Hadamard2_impl(rho0,f2,t2)

# b=Bloch()
# b.add_points([x,y,z])
# b.show()

Bac2=0
Bac1=0.2*2*np.pi
rho2=CNOT_impl(rho1,f1,t1,1)[-1]

Bac1=0
Bac3=0.2*2*np.pi
rho3=CNOT_impl(rho2,f3,t3,3)[-1]

print(rho3)

s0=basis(2,0)
s1=basis(2,1)
ideal=1/np.sqrt(2)*(tensor(s0,s0,s0)+tensor(s1,s1,s1))
print(ket2dm(ideal))
```
